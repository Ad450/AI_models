```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras import layers
import numpy as np
from sklearn.preprocessing import StandardScaler


#  load data
dataset = pd.read_csv("led.csv")
dataset.dropna(inplace=True)

#  select features and target
input = dataset[["GDP", "AdultMortality", "Population"]]

output = dataset[["Lifeexpectancy"]]
x = input.to_numpy()
y = output.to_numpy()

# Scale data
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(x)
y_scaled = scaler_y.fit_transform(y)

#  build model

model = tf.keras.Sequential([
    layers.Input(shape=(3,)),
    layers.Dense(10, activation="relu"),
    layers.Dropout(0.2),
    layers.Dense(1)
])

model.compile(loss = tf.keras.losses.MeanSquaredError(),
                      optimizer = tf.keras.optimizers.Adam())


#  train model
model.fit(X_scaled, y_scaled, batch_size=10, epochs=200)

# test model

test_input = np.array([[600.25921, 260, 387582.0]])
scaled_test_input = scaler_X.transform(test_input)

pred_scaled = model.predict(scaled_test_input)
prediction = scaler_y.inverse_transform(pred_scaled)

prediction




```
